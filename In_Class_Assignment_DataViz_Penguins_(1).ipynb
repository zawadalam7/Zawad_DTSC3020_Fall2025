## 1–3) Scatter plot with color by species
Create a scatter plot with:
- **x =** `bill_length_mm`
- **y =** `bill_depth_mm`
- Color points by **species**

Then add a short comment (2–3 sentences) describing the pattern(s) you see.

```python
import seaborn as sns
import matplotlib.pyplot as plt

penguins = sns.load_dataset("penguins")

plt.figure(figsize=(8, 6))
sns.scatterplot(
  data=penguins,
  x="bill_length_mm",
  y="bill_depth_mm",
  hue="species",
  palette="Set2"
)

plt.title("Bill Length vs. Bill Depth by Penguin Species")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Bill Depth (mm)")
plt.legend(title="Species")
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Your interpretation (2–3 sentences):**

It shows that Adelie penguins have short bills with great depth, Gentoo penguins have longer bills with lesser depth and Chinstrap penguins have both good length and depth on their bills.

## 4) Histogram with two different bin sizes
Plot a histogram of `bill_length_mm` **twice**:
- Once with **10 bins**
- Once with **30 bins**

Then write 1–2 sentences comparing what changes when you change the number of bins.

```python
import seaborn as sns
import matplotlib.pyplot as plt

penguins = sns.load_dataset("penguins")

plt.figure(figsize=(8, 5))
sns.histplot(penguins['bill_length_mm'], bins=10, kde=False, color='skyblue')
plt.title("Histogram of Bill Length (10 bins)")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Frequency")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the penguins dataset
penguins = sns.load_dataset("penguins")

# Plot histogram with 30 bins
plt.figure(figsize=(8, 5))
sns.histplot(penguins['bill_length_mm'], bins=30, kde=False, color='salmon')
plt.title("Histogram of Bill Length (30 bins)")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Frequency")
plt.show()
```

**Your comparison (1–2 sentences):**

The histogram with 10 bins gives a more generatlized view of the bill lengths while the 30 bin histogram provides a lot more detail.
